```python
def vector_vector_multiplication(u, v):
    assert u.shape[0] == v.shape[0]

    n = u.shape[0]
    
    result = 0.0

    for i in range(n):
        result = result + u[i] * v[i]
        
    return result
        
```

```python
vector_vector_multiplication(u, v)
```

```python
u.dot(v)
```
